```python
# we have to convert str list into list using ast
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data.genres=data.genres.apply(convert)
data.keywords=data.keywords.apply(convert)
```

```python
import ast
def convert_cast(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['character'])
            c=c+1
        else:
            break
        
    return l


data.cast=data.cast.apply(convert)
```

```python
import ast
def crew_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l


data.crew=data.crew.apply(crew_director)
```

```python
data.head(2)
```

```python
data.dropna(inplace=True)
```

```python
data['overview'] = data['overview'].apply(lambda x : x.split())
```

```python
data.head()
```

```python
data['cast']=data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['genres']=data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew']=data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords']=data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data.head(1)
```

```python
data['tags']=data['overview']+data['genres']+data['keywords']+data['cast']+data['crew']
```

```python
data.tags=data.tags.apply(lambda x:" ".join(x))
```

```python
movies=data[['movie_id','title','tags']]
movies.head(2)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies.tags=movies.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')

vectors=cv.fit_transform(movies.tags).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#fetching the top 5 similar movies
sorted(enumerate(similarity[0]), key=lambda x: x[1], reverse=True)[1:6]
```

```python
def recomendation(film):
    movie_index= movies[movies.title==film].index[0]
    distances= similarity[movie_index]
    movie_list= sorted(enumerate(distances), key=lambda x: x[1], reverse=True)[1:6]
    
    for i in movie_list:
        print(movies.iloc[i[0]][1])
```
