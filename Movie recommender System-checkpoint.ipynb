- Budget is not an important column.

```python
data=data[['movie_id','title', 'genres', 'overview', 'keywords', 'cast', 'crew']]
data.head(1)
```

```python
# we have to convert str list into list using ast
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data.genres=data.genres.apply(convert)
data.keywords=data.keywords.apply(convert)
```

```python
import ast
def convert_cast(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['character'])
            c=c+1
        else:
            break
        
    return l


data.cast=data.cast.apply(convert)
```

```python
import ast
def crew_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l


data.crew=data.crew.apply(crew_director)
```
